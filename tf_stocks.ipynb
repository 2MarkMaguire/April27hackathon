```python
# You might want to use the following packages
import numpy as np
import os
import tensorflow as tf
tf.logging.set_verbosity(tf.logging.ERROR) #reduce annoying warning messages
from functools import partial

# to make this notebook's output stable across runs
def reset_graph(seed=42):
    tf.reset_default_graph()
    tf.set_random_seed(seed)
    np.random.seed(seed)
```

```python
reset_graph()

n_inputs = 14
n_hidden1 = 7
learning_rate1 = .005
n_outputs = 2

X = tf.placeholder(tf.float32, shape=(None, 14), name="X")
y = tf.placeholder(tf.int32, shape=(None), name="y")

def layer(X, n_neurons, name, activation=None):
    with tf.name_scope(name):
        init = tf.variance_scaling_initializer()
        W = tf.Variable(init)
        b = tf.Variable(tf.zeros([n_neurons]))
```

```python
with tf.name_scope("dnn2"):
    h1 = tf.contrib.layers.variance_scaling_initializer(dtype=tf.float32)
    hidden1 = tf.layers.dense(X, n_hidden1, activation=tf.nn.elu, kernel_initializer=h1, name="hidden1")
    hidden2 = tf.layers.dense(hidden1, n_hidden1, activation=tf.nn.elu, kernel_initializer=h1, name="hidden2")
    hidden3 = tf.layers.dense(hidden2, n_hidden1, activation=tf.nn.elu, kernel_initializer=h1, name="hidden3")
    hidden4 = tf.layers.dense(hidden3, n_hidden1, activation=tf.nn.elu, kernel_initializer=h1, name="hidden4")
    hidden5 = tf.layers.dense(hidden4, n_hidden1, activation=tf.nn.elu, kernel_initializer=h1, name="hidden5")
    hidden6 = tf.layers.dense(hidden5, n_hidden1, activation=tf.nn.elu, kernel_initializer=h1, name="hidden6")
    hidden7 = tf.layers.dense(hidden6, n_hidden1, activation=tf.nn.elu, kernel_initializer=h1, name="hidden7")
    logits = tf.layers.dense(hidden7, n_outputs, name="outputs")
```

```python
with tf.name_scope("loss"):
    pt = tf.nn.sparse_softmax_cross_entropy_with_logits(logits = logits, labels = y)
    loss_result = tf.reduce_mean(pt)
```

```python
with tf.name_scope("train"):
    train_optimizer = tf.train.GradientDescentOptimizer(learning_rate=learning_rate1)
    opt = train_optimizer.minimize(loss_result)
```

```python
with tf.name_scope("eval"):
  #implementation of the evaluation procedure here
    checker = tf.cast(tf.nn.in_top_k(logits, y, 1), tf.float32)
    accuracy = tf.reduce_mean(checker)
```

```python
# Your code goes here
init = tf.global_variables_initializer()
saver = tf.train.Saver()

n_epochs = 15
batch_size = 30

# shuffle_batch() shuffle the examples in a batch before training
def shuffle_batch(X, y, batch_size):
    rnd_idx = np.random.permutation(len(X))
    n_batches = len(X) // batch_size
    for batch_idx in np.array_split(rnd_idx, n_batches):
        X_batch, y_batch = X[batch_idx], y[batch_idx]
        yield X_batch, y_batch
```

```python
with tf.Session() as sess:
  init.run()
  for epoch in range(n_epochs):
        for X_batch, y_batch in shuffle_batch(X_train, y_train, batch_size):
            sess.run(opt, feed_dict={X: X_batch, y: y_batch})
        acc_batch = accuracy.eval(feed_dict={X: X_batch, y: y_batch})
        #acc_val = accuracy.eval(feed_dict={X: X_val, y: y_val})
        print("Iter: ", epoch, "Batch accuracy:", acc_batch)
    
  save_path = saver.save(sess, "./my_dnn_model31.ckpt")
```

```python
with tf.Session() as sess:
    saver.restore(sess, "./my_dnn_model31.ckpt")
    # implementation of the test set evaluation here
    sess.run(opt, feed_dict={X: X_test, y: y_test})
    acc_test = accuracy.eval(feed_dict={X: X_test, y: y_test})
```
